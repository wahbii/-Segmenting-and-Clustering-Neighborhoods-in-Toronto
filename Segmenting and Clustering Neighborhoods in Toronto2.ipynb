```python
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url)[0]
df.head()
```

now we need to clean out data from not assigned value in column 'Borough' 

```python
index=df[df['Borough']=='Not assigned'].index
df.drop(index , inplace=True)
df.reset_index(drop=True)
```

now let's look if theres in 'Neighborhood' rows with NaN  or Not assigned value

```python
df[df['Neighborhood']=='NaN']
```

```python
df[df['Neighborhood']==' Not assigned']
```

as we  see our data is clean 
now we need to replace (/) in Neighborhood column with (,)

```python
import numpy as np
na=[]
x='/'
y=','

for dat in df['Neighborhood']:
    m=dat.replace(x,y)
    na.append(m)
    
new_n=np.array(na)
df.Neighborhood=new_n
```

Now we gonna use file csv witch contain latitude and longitude for each postal code

```python
geo=pd.read_csv('Geospatial_Coordinates.csv')
geo.head()
```

as we see thy have the same nombre of rows 
witch make our work easy  just we ganna ordered the both data frame and concatenate  it (we know that  from the instruction of the assignement the geo contain latitude and longitude  for every postal code in our data  )

```python
geo.sort_values(by='Postal Code',inplace=True)
df.sort_values(by='Postal code',inplace=True)
geo.reset_index(drop=True,inplace=True)
df.reset_index(drop=True,inplace=True)
```
