```python
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url)[0]
df.head()
```

now we need to clean out data from not assigned value in column 'Borough'

```python
index=df[df['Borough']=='Not assigned'].index
df.drop(index , inplace=True)
df.reset_index(drop=True)
```

now let's look if theres in 'Neighborhood' rows with NaN  or Not assigned value

```python
df[df['Neighborhood']=='NaN']
```

```python
df[df['Neighborhood']==' Not assigned']
```

as we see our data is clean now we need to replace (/) in Neighborhood column with (,)

```python
import numpy as np
na=[]
x='/'
y=','

for dat in df['Neighborhood']:
    m=dat.replace(x,y)
    na.append(m)
    
new_n=np.array(na)
df.Neighborhood=new_n
```

```python
import requests
client_id='AU2F2RDPFWG1QIEFTQXLDZB4NBFKQ1150WNDYNESYX0WGGAN'
c_secret='IZSU5IMAELK4TX3JDVNK2JQX11ZXTSE5ZWOKSYMAM4MLYEWD'
version='20180605'

def getvenue(names,latitude,longitude):
    column=['Neighborhood','latiude','longitude', 'venue_name','venue_lat','venue_long','venue_cat']
    datavenue=pd.DataFrame(columns=column)
    radius=500
    limit=100
    for n,lan,lngi in zip(names,latitude,longitude):
        url='http://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(client_id,c_secret,lan,lngi,version,radius,limit)
        resutat=requests.get(url).json()['response']['groups'][0]['items']
        for v in resutat:
            datavenue=datavenue.append({
               'Neighborhood':n,
                'latiude':lan,
                'longitude':lngi,
                'venue_name':v['venue']['name'],
                'venue_lat':v['venue']['location']['lat'],
                'venue_long':v['venue']['location']['lng'],
                'venue_cat' :v['venue']['categories'][0]['name']
         
            },ignore_index=True)
            
            
    return datavenue       
        
```

```python
toront_v=pd.get_dummies(T_venues[['venue_cat']])
toront_v['Neighborhood']=T_venues['Neighborhood']
toront_v.head()
```

make Neighborhood at first column

```python
col=[toront_v.columns[-1]]+list(toront_v.columns[:-1])
toront_v=toront_v[col]
toront_v
```

lest group data by Neighborhood   and see how many venue each Neighborhood

```python
T=toront_v.groupby('Neighborhood').count()
T.head()
```

Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toront_v.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

Cluster Neighborhoods

```python
toronto['Borough'].value_counts()
```

we have 4 borough then lets  do k =4 for clustering

```python
k=4

from sklearn.cluster import KMeans
clust=KMeans(n_clusters=k,random_state=0).fit(toronto_grouped.drop(['Neighborhood'],axis=1))
```

```python
clust.labels_
```

```python
toronto_grouped['labels']=clust.labels_
```

let s add location to Neighborhood in grouped data

```python
locationn=T_venues[['Neighborhood','latiude','longitude']]
```

```python
lat=[]
long=[]

for n in toronto_grouped['Neighborhood']:
        x=locationn[locationn['Neighborhood']==n].latiude
        x=np.array(x)
        y=locationn[locationn['Neighborhood']==n].longitude
        y=np.array(y)
        lat.append(x[0])
        long.append(y[0])
```

```python
toronto_grouped['latitude']=lat
toronto_grouped['longitude']=long
```
